```python
import pandas as pd
import matplotlib.pyplot as plt

file_name = 'cleaned_bank.csv'

f = pd.read_csv(file_name, sep=";")
```

```python
f["age"].plot(kind='hist', bins = 10)
plt.title("Client Age Information")
plt.xlabel("Age")
plt.savefig("./graphs/age.png")
```

```python
f["job"].value_counts().plot(kind='pie', autopct='%.2f')
plt.savefig("./graphs/job.png")
```

```python
f["marital"].value_counts().plot(kind='pie', autopct='%.2f')
plt.savefig("./graphs/marital.png")
```

```python
freq = f["education"].value_counts().values
keys = f["education"].value_counts().axes[0]
plt.bar(keys, freq)
plt.title("Client Education Information")
plt.xlabel("level")
plt.ylabel("# of Clients")
plt.xticks(rotation=90)
plt.savefig("./graphs/education.png", bbox_inches = 'tight')
```

```python
f["default"].value_counts().plot(kind='pie', autopct='%.2f')
plt.savefig("./graphs/default.png")
```

```python
f["housing"].value_counts().plot(kind='pie', autopct='%.2f')
plt.savefig("./graphs/housing.png")
```

```python
f["loan"].value_counts().plot(kind='pie', autopct='%.2f')
plt.savefig("./graphs/loan.png")
```

```python
f["contact"].value_counts().plot(kind='pie', autopct='%.2f')
plt.savefig("./graphs/contact.png")
```

```python
freq = f["month"].value_counts().values
keys = f["month"].value_counts().axes[0]
plt.bar(keys, freq)
plt.title("Last Contact Month of Year")
plt.xlabel("month")
plt.ylabel("# of clients contacted")
plt.xticks(rotation=90)
plt.savefig("./graphs/month.png", bbox_inches = 'tight')
```

```python
freq = f["day_of_week"].value_counts().values
keys = f["day_of_week"].value_counts().axes[0]
plt.bar(keys, freq)
plt.title("Last Contact Day of Week")
plt.xlabel("day")
plt.ylabel("# of clients contacted")
plt.xticks(rotation=90)
plt.savefig("./graphs/day.png", bbox_inches = 'tight')
```

```python
f["duration"].plot(kind='box')
plt.savefig("./graphs/duration.png")
```

```python
f["campaign"].plot(kind='box')
plt.savefig("./graphs/campaign.png")
```

```python
f["pdays"].value_counts(sort=False).plot(kind='bar')
plt.title("Days Passed Since Contact From Prev Campaign")
plt.xlabel("# of days")
plt.ylabel("# of clients contacted")
plt.savefig("./graphs/pdays.png", bbox_inches = 'tight')
```

```python
f["previous"].plot(kind='hist')
plt.title("Contacts Performed Before Campaign")
plt.xlabel("# of contacts performed")
plt.ylabel("# of clients contacted")
plt.savefig("./graphs/previous.png")
```

```python
f["poutcome"].value_counts().plot(kind='pie', autopct='%.2f')
plt.savefig("./graphs/poutcome.png")
```

```python
f["emp.var.rate"].plot(kind='hist')
plt.title("Employement Variation")
plt.xlabel("Emp.var.rate")
plt.ylabel("# of clients")
plt.savefig("./graphs/emp_var.png")
```

```python
f["cons.price.idx"].plot(kind='hist')
plt.title("Consumer Price Index")
plt.xlabel("cons.price.idx")
plt.ylabel("# of clients")
plt.savefig("./graphs/cons_price.png")
```

```python
f["cons.conf.idx"].plot(kind='hist')
plt.title("Consumer Confidence Index")
plt.xlabel("cons.conf.idx")
plt.ylabel("# of clients")
plt.savefig("./graphs/cons_conf.png")
```

```python
f["euribor3m"].plot(kind='hist')
plt.title("Euribor 3 Month Rate")
plt.xlabel("rate")
plt.ylabel("# of clients")
plt.savefig("./graphs/euribor3m.png")
```

```python
f["nr.employed"].plot(kind='hist')
plt.title("Number of Employees")
plt.ylabel("# of clients")
plt.savefig("./graphs/nr_emp.png")
```

```python
f["y"].value_counts().plot(kind='pie', autopct='%.2f')
plt.title("Clients Subscribed")
plt.savefig("./graphs/subscribed.png")
```

```python
f.boxplot(column='age', by='y')
plt.ylabel('age')
plt.xlabel('client subscription')
plt.savefig("./graphs/age_target.png")
```

```python
f.boxplot(column='cons.price.idx', by='y')
plt.ylabel('Consumer Price Index')
plt.xlabel('client subscription')
plt.savefig("./graphs/cons_price_target.png")
```

```python
f.boxplot(column='euribor3m', by='y')
plt.ylabel('Euribor 3 Month Rate')
plt.xlabel('client subscription')
plt.savefig("./graphs/euribor3m_target.png")
```

```python
from pandas.plotting import scatter_matrix
colors_palette = {'yes': 'red', 'no': 'green'}
colors = [colors_palette[i] for i in f["y"]]

axes = scatter_matrix(f, alpha=0.2, figsize=(35,35), c=colors, diagonal='hist')

#y ticklabels
[plt.setp(item.yaxis.get_majorticklabels(), 'size', 15) for item in axes.ravel()]
#x ticklabels
[plt.setp(item.xaxis.get_majorticklabels(), 'size', 15) for item in axes.ravel()]
#y labels
[plt.setp(item.yaxis.get_label(), 'size', 23) for item in axes.ravel()]
#x labels
[plt.setp(item.xaxis.get_label(), 'size', 23) for item in axes.ravel()]

plt.savefig("./graphs/scatter_matrix.png")
```
